```python
# 用户登录风险打分
logins = [
    {"user_id": 101, "device": "mobile", "city": "Beijing",  "fail_count": 0},
    {"user_id": 102, "device": "pc",     "city": "Shanghai", "fail_count": 1},
    {"user_id": 103, "device": "mobile", "city": "Other",    "fail_count": 3},
    {"user_id": 104, "device": "pc",     "city": "Other",    "fail_count": 5},
    {"user_id": 105, "device": "mobile", "city": "Beijing",  "fail_count": 2},
]

# 小函数：只接收“一个值”，函数内部没有 tx，用参数就行
def device_risk(device):
    risk = 0
    if device == "mobile":
        risk += 5
    else:
        risk += 10
    return risk

def city_risk(city):
    risk = 0
    if city == "Beijing":
        risk += 0
    elif city == "Shanghai":
        risk += 5
    else:
        risk += 10
    return risk

def failcount_risk(fail_count):
    risk = 0
    if fail_count == 0:
        risk += 0
    elif 1 <= fail_count <= 3:
        risk += 10
    else:
        risk += 30
    return risk

# 总风险：接收整条记录 tx，从 tx 里取值，再调用上面三个函数
def calc_login_risk(tx):
    device = tx["device"]
    city = tx["city"]
    fail_count = tx["fail_count"]

    risk = 0
    risk += device_risk(device)
    risk += city_risk(city)
    risk += failcount_risk(fail_count)
    return risk

def exercise3():
    for tx in logins:
        total_risk = calc_login_risk(tx)
        print(
            f"用户 {tx['user_id']}: device={tx['device']}, city={tx['city']}, "
            f"fail_count={tx['fail_count']}, risk={total_risk}"
        )

exercise3()
        
        
        
        
        
```

```python
# 借款申请风控 + 高风险筛选
applications = [
    {"app_id": 1, "amount": 3,   "term_months": 6,  "has_guarantor": True},
    {"app_id": 2, "amount": 10,  "term_months": 12, "has_guarantor": False},
    {"app_id": 3, "amount": 25,  "term_months": 24, "has_guarantor": True},
    {"app_id": 4, "amount": 8,   "term_months": 12, "has_guarantor": False},
    {"app_id": 5, "amount": 50,  "term_months": 24, "has_guarantor": False},
]

def amount_risk(amount):
    risk = 0
    if amount <5:
        risk += 5
    elif 5 <= amount < 20:
        risk += 15
    else:
        risk += 30
    return risk

def term_risk(term_months):
    risk = 0
    if term_months <= 6:
        risk += 0
    elif 6 < term_months <= 12:
        risk += 10
    else:
        risk += 20
    return risk

def guarantor_risk(has_guarantor):
    risk = 0
    if has_guarantor is True:
        risk += 0
    else:
        risk += 15
        
    return risk
        
def calc_app_risk(tx):
    amount = tx["amount"]
    term_months = tx["term_months"]
    has_guarantor = tx["has_guarantor"]
    
    risk = 0
    risk += amount_risk(amount)
    risk += term_risk(term_months)
    risk += guarantor_risk(has_guarantor)
    
    return risk

def exercise5():
    # 第一步：全部打印
    for tx in applications:
        total_risk = calc_app_risk(tx)
        print(f"申请={tx['app_id']}, amount={tx['amount']}, term_months={tx['term_months']}, has_guarantor={tx['has_guarantor']}, risk={total_risk}")
        
    
    print("---")  
    # 第二步：只打印 risk >= 40 的
    for tx in applications:
        total_risk = calc_app_risk(tx)
        if total_risk >= 40:
            print(f"申请={tx['app_id']}, amount={tx['amount']}, term_months={tx['term_months']}, has_guarantor={tx['has_guarantor']}, risk={total_risk}")
    
exercise5()
```

```python
# 用“规则表”重写金额打分
amount_rules = [
    {"name": "小于等于0", "condition": lambda x: x <= 0,              "score": 100},
    {"name": "500到5000", "condition": lambda x: 500 < x <= 5000,     "score": 30},
    {"name": "大于5000",  "condition": lambda x: x > 5000,            "score": 60},
    {"name": "临界值",    "condition": lambda x: x in [499, 500, 501], "score": 20},
    {"name": "偶数加分",  "condition": lambda x: x % 2 == 0,          "score": 5},
]

def calc_risk_by_rules(amount, rules):
    risk = 0
    for rule in rules:
        cond = rule["condition"]   # 取出这一条规则的“条件函数”
        score = rule["score"]      # 取出这一条规则的“加几分”
        if cond(amount):           # 相当于：如果这条规则满足
            risk += score          # 就把对应分数加上
    return risk

def calc_risk2(amount):
    return calc_risk_by_rules(amount, amount_rules)

def exercise10():
    amounts = [12, 880, 55, 3000, 499, 0, -20, 6000]
    for amt in amounts:
        old = calc_risk(amt)    # 你之前写的 if/elif 版本
        new = calc_risk2(amt)   # 现在的规则表版本
        print(f"amount={amt}, old={old}, new={new}")

exercise10()
```
