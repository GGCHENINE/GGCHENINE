```python
frame = pd.DataFrame({"data1": np.random.standard_normal(1000),
                       "data2": np.random.standard_normal(1000)})
frame
```

```python
quartilies = pd.cut(frame["data1"],4)
quartilies
```

```python
def get_stats(group):
    return pd.DataFrame(
        {'min':group.min,'max':group.max,
        'count':group.count(),
        'mean':group.mean()}
    )
grouped = frame.groupby(quartilies)
grouped.apply(get_stats)
```

```python
# 等价于
grouped.agg(['min','max','count','mean'])
```

```python
quartiles_samp = pd.qcut(frame["data1"], 4, labels=False)
# 不返回区间标签，返回编号
quartiles_samp.head()
```

```python
grouped = frame.groupby(quartiles_samp)
grouped.apply(get_stats)
```

```python
states = ["Ohio", "New York", "Vermont", "Florida",
           "Oregon", "Nevada", "California", "Idaho"]

group_key = ["East", "East", "East", "East",
              "West", "West", "West", "West"]

data = pd.Series(np.random.standard_normal(8), index=states)

data
```

```python
data[["Vermont", "Nevada", "Idaho"]] = np.nan
```

```python
data
```

```python
data.groupby(group_key).size()
```

```python
data.groupby(group_key).count()
```

```python
def fill_mean(group):
    return group.fillna(group.mean())
data.groupby(group_key).apply(fill_mean)
# 按照均值补充缺失值
```

```python
fill_values = {"East":0.5,"West":1}
def fill_func(group):
    return group.fillna(fill_values[group.name])
data.groupby(group_key).apply(fill_func)
# 按照组内部name，设置每组缺失值的预设值然后填充
```

```python
# 加权平均
df = pd.DataFrame({"category": ["a", "a", "a", "a",
                                 "b", "b", "b", "b"],
                    "data": np.random.standard_normal(8),
                    "weights": np.random.uniform(size=8)})
df
```

```python
grouped = df.groupby("category")
def get_wavg(group):
    return np.average(group['data'],weights=group['weights'])
grouped.apply(get_wavg)
```

```python
def spx_corr(group):
    return group.corrwith(group['SPX'])
# 对 group 这个分组数据里的每一列，都去和 group['SPX'] 做相关系数计算。
```

```python
reteruns = close_px.pct_change().dropna()
```

```python
def get_year(x):
    return x.year
```

```python
by_year = reteruns.groupby(get_year)
```

```python
df = pd.DataFrame({'key': ['a', 'b', 'c'] * 4,
                    'value': np.arange(12.)})
df
```

```python
g = df.groupby('key')['value']
```

```python
g.mean()
```

```python
def get_mean(group):
    return group.mean()

g.transform(get_mean)
# 对 g 的每个分组求均值，然后把结果“按原来的行形状展开返回”
# transform 会把“每组算出来的结果”再对齐回原来的每一行。
```

```python
g.transform('mean')
```

```python
def times_two(group):
    return group * 2

g.transform(times_two)
```

```python
def get_rank(group):
    return group.rank(ascending=False)

g.transform(get_rank)
# 按数据大小排序
```

```python
def normalize(x):
    return (x-x.mean()) / x.std()
# 把不同量纲、不同大小范围的数据拉到可比较的尺度上
g.transform(normalize)
# 对每个分组分别做标准化,每个组内部都减去本组均值、再除以本组标准差
```

```python
g.apply(normalize)
```

```python
g.transform('mean')
```
