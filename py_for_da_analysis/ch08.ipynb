合并数据集

```python
df1 = pd.DataFrame({"key":['b','b','a','c','a','a','b'],
                    'data1':pd.Series(range(7),dtype="Int64")})
df2 = pd.DataFrame({"key":['a','b','d'],
                    "data2":pd.Series(range(3),dtype="Int64")})
display(df1,df2)
```

```python
pd.merge(df1,df2)
# pandas.merge 会使用重叠的列名作为键
```

```python
pd.merge(df1,df2,on="key")
# 建议明确指定列后连接
```

```python
df3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
                     "data1": pd.Series(range(7), dtype="Int64")})

df4 = pd.DataFrame({"rkey": ["a", "b", "d"],
                     "data2": pd.Series(range(3), dtype="Int64")})
display(df3,df4)
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
# 抹去了c，d值和相关数据，pandas.merge默认为键的交集
```

```python
display(df1,df2)
pd.merge(df1,df2,how="outer")
```

```python
pd.merge(df3,df4,left_on="lkey",right_on="rkey",how="outer")
```

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                     "data1": pd.Series(range(6), dtype="Int64")})

df2 = pd.DataFrame({"key": ["a", "b", "a", "b", "d"],
                     "data2": pd.Series(range(5), dtype="Int64")})
display(df1,df2)
```

```python
pd.merge(df1,df2,on="key",how="left")
# 多对多合并会形成匹配键的笛卡尔积
```

```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                      "key2": ["one", "two", "one"],
                      "lval": pd.Series([1, 2, 3], dtype='Int64')})

right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                       "key2": ["one", "one", "one", "two"],
                       "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
display(left,right)
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
pd.merge(left,right,on='key1')
```

```python
pd.merge(left, right, on="key1", suffixes=("_left", "_right"))
# 当两个表里有同名列时，merge 后给重名列自动加后缀，避免冲突
```

基于索引合并


```python
left1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"],
                       "value": pd.Series(range(6), dtype="Int64")})
right1 = pd.DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
display(left1,right1)
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)
# 将右表的索引当成键与左表匹配，未写how默认inner，left_on指定某列作为键
```

```python
lefth = pd.DataFrame({"key1": ["Ohio", "Ohio", "Ohio",
                                "Nevada", "Nevada"],
                       "key2": [2000, 2001, 2002, 2001, 2002],
                       "data": pd.Series(range(5), dtype="Int64")})
```

```python
display(lefth)
```

```python
righth_index = pd.MultiIndex.from_arrays(
    [
        ['Nevada','Nevada','Ohio','Ohio','Ohio','Ohio'],
        [2001,2000,2000,2000,2001,2002]
    ]
)
righth = pd.DataFrame({"event1":pd.Series([0,2,4,6,8,10],dtype="Int64",
                                          index=righth_index),
                                          "event2": pd.Series([1, 3, 5, 7, 9, 11], dtype="Int64",
                                        index=righth_index)})
```

```python
display(lefth,righth)
```

```python
pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True,
         )
```

```python
pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True,
         how = "outer")
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                      index=["a", "c", "e"],
                      columns=["Ohio", "Nevada"]).astype("Int64")
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                       index=["b", "c", "d", "e"],
                       columns=["Missouri", "Alabama"]).astype("Int64")
display(left2,right2)
```

```python
pd.merge(left2,right2,left_index=True,right_index=True)
# 使用两边的索引合并
```

```python
left2.join(right2,how="outer")
# join默认左连接
```

```python
display(left1,right1)
left1.join(right1,on="key")
# join也支持调用DataFrame的某列做连接
```

```python
another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                        index=["a", "c", "e", "f"],
                        columns=["New York", "Oregon"])
display(another)
```

```python
display(left2,right2)
```

```python
left2.join([right2,another])
# 把 right2 和 another 这两个表，一起按索引连接到 left2 上
```

```python
left2.join([right2,another],how="outer")
```

沿着轴连接

```python
arr = np.arange(12).reshape((3,4))
arr
```

```python
np.concatenate([arr,arr],axis=1)
# 在numpy中，只认轴编号。
# 二维数组中：
# axis=0 表示行轴
# axis=1 表示列轴
```

```python
s1 = pd.Series([0, 1], index=["a", "b"], dtype="Int64")

s2 = pd.Series([2, 3, 4], index=["c", "d", "e"], dtype="Int64")

s3 = pd.Series([5, 6], index=["f", "g"], dtype="Int64")

display(s1,s2,s3)
```

```python
pd.concat([s1,s2,s3])
# 把 s1、s2、s3 这几个对象沿着默认方向拼接起来。
# 如果 s1、s2、s3 是 Series，默认是按 行方向 拼，也就是上下接起来
```

```python
pd.concat([s1,s2,s3],axis=1)
```

```python
result = pd.concat([s1,s2,s3],keys=['one','two','three'])
result
# 通过keys在连接轴上加一个层次索引
```

重叠数据的合并

```python
a = pd.Series([np.nan, 2.5, 0.0, 3.5, 4.5, np.nan],
               index=["f", "e", "d", "c", "b", "a"])

b = pd.Series([0., np.nan, 2., np.nan, np.nan, 5.],
               index=["a", "b", "c", "d", "e", "f"])
display(a,b)
```

```python
np.where(pd.isna(a),b,a)
# 如果 a 这个位置是缺失值，就用 b 的值；否则就用 a 自己的值
# np.where(condition, x, y)
```

```python
a.combine_first(b)
# 优先用 a 的值，如果 a 缺失，就用 b 对应位置的值补上。
```

```python
df = pd.DataFrame({"key": ["foo", "bar", "baz"],
                   "A": [1, 2, 3],
                    "B": [4, 5, 6],
                    "C": [7, 8, 9]})
```

```python
df
```

```python
melted = pd.melt(df,id_vars='key')
melted
# 把 df 从“宽表”变成长表，并且把 key 这一列保留下来作为标识列
```

```python
reshaped = melted.pivot(index="key", columns="variable",
                         values="value")
reshaped
```

```python
reshaped.reset_index()
# pivot 操作会根据用作行标签的列创建索引，因此我们可能需要使用 reset_index 将数据移回列中
```

```python
pd.melt(df, id_vars="key", value_vars=["A", "B"])
#  保留 key，只把 A 和 B 变成长表
```

```python
pd.melt(df,value_vars=['A','B','C'])
# 不指定分组标识符
```
