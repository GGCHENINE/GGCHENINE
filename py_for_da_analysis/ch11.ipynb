```python
p = pd.Period("2012Q4",freq="Q-JAN")
p
# Q-JAN 里的 JAN 意思不是“从 1 月开始”，而是“这个财年在 1 月结束”
```

```python
p.asfreq("D",how="start")
# 把这个“季度 period”转换成“天 period”，并取这个季度的起始那一天
```

```python
p4pm = (p.asfreq("B", how="end") - 1).asfreq("min", how="start") + 16 * 60
# business day 取该季度的最后对应工作日
# 找到季度结束时对应的最后一个工作日
# 再往前退一个工作日
# 取那天的 00:00
# 再加 16 小时
# 得到那天下午 4 点
```

```python
p4pm
```

```python
p4pm.to_timestamp()
# 将时间段变成时间点 
```

```python
dates = pd.date_range("2000-01-01", periods=12, freq="min")

ts = pd.Series(np.arange(len(dates)), index=dates)

ts
```

```python
ts.resample("5min").sum()
```

```python
ts.resample("5min",closed="right").sum()
# 把时间序列 ts 按 5分钟 一组重新分箱，然后对每组求和
# closed="right"左开右闭 （0:00，5:00]因此0:00不计入该组
```

```python
from pandas.tseries.frequencies import to_offset

result = ts.resample("5min",closed="right",label="right").sum()
result.index = result.index + to_offset("-1s")

result
```

```python
# 开盘价-最高价-最低价-收盘价重采样
ts = pd.Series(np.random.permutation(np.arange(len(dates))),index =dates)
ts.resample("5min").ohlc()
# permutation将数随机打乱,ohlc()时间段的第一个值，最大值最小值和最后一个值
```

```python
frame = pd.DataFrame(np.random.standard_normal((24, 4)),
                      index=pd.period_range("1-2000", "12-2001",
                                            freq="M"),
                      columns=["Colorado", "Texas", "New York", "Ohio"])
frame.head()
```

```python
annual_frame = frame.resample("Y-DEC").mean()
annual_frame
```

```python
annual_frame.resample("Q-DEC",convention="end").ffill()
```

```python
annual_frame.resample("Q-DEC",convention="end").asfreq()
# convention="end" 把年转换为季度时把这一年的值放到该年的最后一个季度
# 这里表示只是按新频率展开，不做平均、求和这些聚合
```
