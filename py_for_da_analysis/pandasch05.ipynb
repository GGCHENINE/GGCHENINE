```python
def f1(x):
    return x.max() - x.min()

frame.apply(f1)
```

函数 f （计算一个 Series 的最大值和最小值之差）对 frame 中的每一列都调用一次。结果是一个 Series，其索引为 frame 的列。

```python
display(frame)
frame.apply(f1,axis="columns")
# 按行最大减最小
```

```python
def f2(x):
    return pd.Series([x.min(),x.max()],index=["mix","max"])
display(frame)
frame.apply(f2)
```
