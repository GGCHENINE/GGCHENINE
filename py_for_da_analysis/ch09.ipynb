```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax3.plot(np.random.standard_normal(50).cumsum(),
         color="black",
         linestyle="dashed")
# .cumsum()对50个随机数做随机和
# linestyle="dashed"线型为虚线
ax1.hist(np.random.standard_normal(100),bins=20,color="black",alpha=0.3);
ax2.scatter(np.arange(30),np.arange(30) + 3*np.random.standard_normal(30));
plt.show()
# hist直方图，bins分组，alpha为透明度
# scatter散点图 np.arrange(30)将横坐标设置为0-29 纵坐标设置为0-29+随机扰动项
# 生成一组横纵坐标数据
# 调用 scatter() 把这些数据画成散点图 
```

fig.add_subplot()：一个一个加子图
plt.subplots()：一次性创建整张图和一组子图
1. fig.add_subplot()
先有 fig，再往里面手动加子图。
例如：
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
意思是：

先创建空画布
再逐个添加子图

2. plt.subplots()
一步就把画布和子图都建好。

例如：

fig, axes = plt.subplots(2, 2)
意思是：

创建一个 fig
同时创建 4 个子图
返回成一个 axes 数组


```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.standard_normal(30).cumsum(),
        color="black",
        linestyle="--",marker="o")
# 折线图
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()
ax.plot(data,color="black",linestyle="--",
        label="default"); #将该线标签为default
ax.plot(data,color="black",linestyle="--",
        drawstyle="steps-post",label="steps-post")
# 阶梯状线
ax.legend()
# legend()显示图例
```

```python
fig , ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum());
# plt.subplots() 会同时创建：
# fig：整张图 ax：子图
ticks= ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'],
                            rotation=30,fontsize=8)

ax.set_xlabel("TEST")
ax.set_title("TEST")
ax.set(title="TEST",xlabel="SECRET") 
# set_xticks:matplotlib 在数据范围内放置刻度的位置
# set_xticklabels:自定义x标签，旋转30度，文字大小为8
```

```python
from datetime import datetime
fig , ax = plt.subplots()
data = pd.read_csv('examples/spx.csv',index_col=0,parse_dates=True)
spx = data["SPX"]
# 取出spx这一列
spx.plot(ax=ax,color="black")
# 将spx时间序列画到ax上
crisis_data = [
    (datetime(2007,10,11),"Peak of bull market"),
    (datetime(2008,3,12),"Bear Stearns Fails"),
    (datetime(2008,9,15),"Lehman Bankruptcy")
]
# 注释点
for date , label in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date)+75),
                xytext=(date,spx.asof(date)+225),
# xy=：箭头指向的位置 xytext=：文字放置的位置 arrowprops=：箭头样式
                arrowprops=dict(facecolor="black",headwidth=4,width=2,
                                headlength=4),
                horizontalalignment="left",verticalalignment="top")
ax.set_xlim(["1/1/2007",'1/1/2011'])
ax.set_ylim([600,1800])
# 限制坐标范围
ax.set_title("Important dates in the 2008-2009 financial crisis")
plt.show()
    
```

```python
fig ,ax = plt.subplots()
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color="black",alpha=0.3)
circ = plt.Circle((0.7,0.2),0.15,color="blue",alpha=0.3)
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],
                    color="green",alpha=0.6)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
# ax.add_patch() 一次只接收 一个 patch 对象，不能一次传多个。
# 把“一个图形补丁对象”加到坐标轴里。
plt.show()
```

```python
tips = pd.read_csv('examples/tips.csv')
tips.head()
```

```python
party_counts = pd.crosstab(tips['day'],tips['size'])
# 交叉表，统计两个或多个分类变量的频数分布
party_counts
```

```python
party_counts = party_counts.loc[:,2:5]
party_counts
```

```python
party_pcts = party_counts.div(party_counts.sum(axis="columns"),
                              axis=0)
#  party_counts 的每一行都除以该行总和
party_pcts
```

```python
party_pcts.plot.barh()
```

```python
macro = pd.read_csv("examples/macrodata.csv")

data = macro[["cpi", "m1", "tbilrate", "unemp"]]

trans_data = np.log(data).diff().dropna()
# 取自然对数做一阶差分（每列）
display(np.log(data))
trans_data.tail()
```

```python
ax = sns.regplot(x="m1",y="unemp",data=trans_data)
# 使用 seaborn 的 regplot 方法，它会生成散点图并拟合一条线性回归线
ax.set_title("Changes in log(m1) bersus log(unemp)")
```

```python
sns.pairplot(trans_data,diag_kind="kde",plot_kws={"alpha":0.2})
# 散点图矩阵
```

```python
sns.catplot(x="day", y="tip_pct", hue="time", col="smoker",
             kind="bar", data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x="tip_pct", y="day", kind="box",
             data=tips[tips.tip_pct < 0.5])
# 箱线图
```
