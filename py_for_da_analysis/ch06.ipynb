```python
chunker=pd.read_csv('examples/ex6.csv',chunksize=1000)
tot = pd.Series([],dtype='int64')
for piece in chunker:
    tot = tot.add(piece['key'].value_counts(),fill_value=0)

tot = tot.sort_values(ascending=False)
tot[:10]
```

这段代码的核心思想是：不一次性把整个大文件读进内存，而是分块读取，每读一块就统计一次 key 列频数，最后把所有块的统计结果累加起来。

读取Excel文件
pandas 也支持使用 pandas.ExcelFile 类或 pandas.read_excel 函数读取存储在 Excel 2003（及更高版本）文件中的表格数据。内部，这些工具分别使用 xlrd 和 openpyxl 插件包来读取旧式 XLS 文件和较新的 XLSX 文件。这些必须使用 pip 或 conda 单独于 pandas 进行安装

```python
import pandas as pd
xlsx = pd.ExcelFile("examples/ex1.xlsx")
xlsx.sheet_names
```

```python
xlsx.parse(sheet_name='Sheet1')
# 从 Excel 文件对象里，把 Sheet1 这个工作表读成 DataFrame
```

```python
frame = pd.read_excel("examples/ex1.xlsx",sheet_name="Sheet1")
frame
```

```python
frame = pd.DataFrame({'a':np.random.standard_normal(100)}) #生成100个标准正态分布随机数，存入列名为a的DataFrame中
store = pd.HDFStore('examples/mydata.h5') #创建一个HDF5文件
store['obj1'] = frame #将frame存进去，键名为obj2
store['obj1_col'] = frame['a']
# 把 frame['a'] 这一列单独存进去，键名是 "obj1_col"
store
```

在 pandas 里，字典的：

key -> 列名
value -> 这一列的数据

```python
store['obj1']
```

```python
store.put('obj2',frame,format='table')
store.select('obj2',where=['index>=10 and index <= 15'])
```

```python
import sqlite3
```

```python
query = """
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
 c REAL,        d INTEGER
);"""
```

```python
con = sqlite3.connect("mydata.sqlite")
con.execute(query)
```

```python
con.commit
```

```python
data = [('Atlanta','Georgia',1.25,6),
        ("Tallahassee","Florida",2.6,3),
        ("Sacramento","California",1.7,5)]
stmt = "INSERT INTO test VALUES(?,?,?,?)"
# SQL 插入语句
con.executemany(stmt,data)
# 批量执行插入
con.commit()
```

```python
cursor = con.execute("SELECT * FROM test")
rows = cursor.fetchall()
rows
```
