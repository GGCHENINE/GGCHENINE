计算指示/虚拟变量

```python
df = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                    "data1": range(6)})
```

```python
df
```

```python
pd.get_dummies(df["key"],dtype=float)
# 把 df["key"] 这一列的分类值转换成哑变量矩阵
# 把文字类别变成计算机更容易处理的数字形式
```

```python
dummies = pd.get_dummies(df["key"],prefix="key",dtype=float)
df_with_dummy = df[['data1']].join(dummies)
df_with_dummy
# 先把 key 列做成哑变量，再把这些哑变量列拼接到 data1 那一列旁边
# 通过prefix="key",列名变成key_a key_b key_c
# df[['data1]]取出原表的data1，使用双中括号，结果还是DataFrame而不是Series
# 通过join()横向拼接
```

正则表达式

```python
import re
```

```python
text = "foo.   bar\t.   baz.   \tqux"
text
```

```python
re.split(r"\s+",text)
```

```python
regex = re.compile(r"\s+")
regex.split(text)
# 先编译一个“匹配一个或多个空白字符”的正则表达式，再用它去切分字符串 text
```

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com"""
pattern = r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}"
```

```python
regex = re.compile(pattern,flags=re.IGNORECASE)
regex.findall(text)
# flags= 匹配是额外采用的规则
```

```python
m = regex.search(text)
m
```

```python
print(regex.sub("REDACTED",text))
# 将匹配值替换为指定值
```

```python
pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern,flags = re.IGNORECASE)
```

```python
m = regex.match("wesm@bright.net")
m
```

```python
m.groups()
# match()从字符串开头开始匹配，group()则从匹配结果中取出匹配内容
```

```python
regex.findall(text)
```

```python
print(regex.sub(r"Username: \1, Domain: \2, Suffix: \3", text))
# 通过\1,\2访问匹配的分组
```

对分类数据的计算

```python
rng = np.random.default_rng(seed=12345)
# random number generator
draws = rng.standard_normal(1000)
draws[:5]
```

```python
bins = pd.qcut(draws,4)
```

```python
bins
```

```python
bins = pd.qcut(draws,4,labels=['Q1','Q2','Q3','Q4'])
bins
```

```python
bins.codes[:10]
# 前十个值落在哪号箱
```

```python
bins =pd.Series(bins,name='quartitle')
results = (pd.Series(draws).groupby(bins).agg(['count','min','max']).reset_index())
# 按分箱区间对 draws 分组，统计每个区间的数量、最小值和最大值
results
```

```python
results['quartitle']
```
